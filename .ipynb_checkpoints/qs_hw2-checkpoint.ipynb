```python
# 20. Valid Parentheses
class Solution:
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        map_list = {")":"(","]":"[","}":"{"}
        res = ""
        for char in s:
            if char not in map_list:
                res = res+char
            elif len(res)==0:
                return False
            elif res[-1]!=map_list[char]:
                return False
            else:
                res=res[:-1]

        if res:
            return False
        else:
            return True
```

```python
# 32. Longest Valid Parentheses
class Solution:
    def longestValidParentheses(self, s):
        """
        :type s: str
        :rtype: int
        """
        maxLen = 0
        stack = [-1]
        for i in range(len(s)):
            if s[i]=="(":
                stack.append(i)
            elif len(stack)>0:
                stack.pop()
                if len(stack)>0:
                    maxLen = max(maxLen,i-stack[-1])
                else:
                    stack.append(i)
        return maxLen
```

```python
# 503. Next Greater Element II
class Solution:
    def nextGreaterElements(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        res = [-1] * len(nums)
        stack = []
        for _ in range(2):
            for i in range(len(nums)):
                while stack and nums[stack[-1]]<nums[i]:
                    res[stack.pop()] = nums[i]
                else:
                    stack.append(i)
        return res
        
```
